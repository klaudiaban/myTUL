# Info about the details
* box_loss - Bounding box accuracy (localization) [Should decrease]
* cls_loss	Class prediction loss [Should decrease]
* dfl_loss	Distribution focal loss (box shape/confidence) [Should decrease]
* P	Precision: how many predicted boxes are correct [Close to 1 (100%)]
* R	Recall: how many ground truths are found [Close to 1 (100%)]
* mAP@0.5	Mean Average Precision at IoU=0.5 (core score) [> 0.5 is decent]
* mAP@.5:.95	mAP averaged across IoU thresholds (harder) [> 0.3–0.5 for good]
